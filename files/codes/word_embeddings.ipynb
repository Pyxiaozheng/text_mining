### 1. 加载和预处理新闻数据

```python
from nltk.corpus.reader import CategorizedTaggedCorpusReader
```

```python
creader = CategorizedTaggedCorpusReader('./cn_news_tagged/', '.*',cat_pattern = r'(.+)/.+txt')
```

```python
import re
import string
```

```python
zh_char = re.compile(r'[\u4e00-\u9fa5]')
```

```python
with open('./stopwords.txt') as f:
    stopwords = f.read().strip().split()
    
domain_stopwords = ['记者', '报道']
```

```python
cn_filter = lambda item: len(zh_char.findall(item)) > 1 and item not in (stopwords + domain_stopwords) and not set(item) & set(string.punctuation)
```

```python
tokenized_docs = []
for file_id in creader.fileids():
    doc = creader.words(fileids=[file_id])
    clean_doc = filter(cn_filter, doc)
    tokenized_docs.append(clean_doc)
```

```python
len(tokenized_docs)
```

### 2. 训练Word2Vec模型

```python
import gensim
from gensim.models import Word2Vec
```

```python
# 为了加快本机处理速度，只使用语料库中1000条新闻作为训练数据
# 为了获取更好的词向量表示，请尽可能使用更多的语料
tokenized_data = [list(line) for line in tokenized_docs[:1000]]
```

```python
# 训练 Word2Vec 模型
# Gensim 的 Word2Vec 实现默认使用 Skip-gram 架构
# 如果要使用 CBOW（Continuous Bag-of-Words）架构，需要在创建 Word2Vec 模型时将 sg 参数设置为 0。
model = Word2Vec(sentences=tokenized_data, vector_size=100, window=5, min_count=1, workers=4)

# 保存训练好的模型
model.save("./cn_news_word2vec.model")
```

```python
# 加载模型
loaded_model = Word2Vec.load("./cn_news_word2vec.model")
```

```python
# 查看词表
# {word: index}
vocab_dict = loaded_model.wv.key_to_index
```

```python
# 获取某个词语的向量
loaded_model.wv["中国"]
```

### 3. 根据余弦相似度寻找语义相似的词语

```python
# 查找与给定单词最相似的词
word = "中国"
num_similar_words = 5  # 指定要返回的相似词数量
similar_words = loaded_model.wv.most_similar(word, topn=num_similar_words)

print(f"Words most similar to '{word}':")
for similar_word, similarity in similar_words:
    print(f"{similar_word}: {similarity}")
```

### 4. 可视化词向量

```python
import numpy as np
import matplotlib.pyplot as plt
from gensim.models import Word2Vec
from sklearn.manifold import TSNE

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 加载模型
model = Word2Vec.load("./cn_news_word2vec.model")

# 获取词汇表中的词向量
# 选择前 N 个最频繁的单词
# 为提高运行速度，只绘制100个词汇的二维可视化图
N = 100  # 要可视化的单词数
most_frequent_words = sorted(model.wv.key_to_index, key=model.wv.key_to_index.get)[:N]

labels = []
vectors = []
for word in most_frequent_words:
    labels.append(word)
    vectors.append(model.wv[word])

# 使用 t-SNE 降维
tsne = TSNE(n_components=2, random_state=42)
reduced_vectors = tsne.fit_transform(vectors)

# 绘制降维后的词向量
plt.figure(figsize=(12, 8))
for i, label in enumerate(labels):
    x, y = reduced_vectors[i, :]
    plt.scatter(x, y)
    plt.annotate(label, xy=(x, y), xytext=(5, 2), textcoords="offset points", ha="right", va="bottom")

plt.show()
```
